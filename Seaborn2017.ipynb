```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=20)
#kde=False make the line dispear #kernell density estimation
#bins
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
#kind = 'hex'
#kind ='reg' 'kde'
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')#pari compare all the values
#use hue to show the difference on factors
#palette shows the styles
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
#kdeplot:if you only need kde plot
sns.kdeplot(tips['total_bill'])
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,cmap='coolwarm')
#most popular travel month is summer
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm')#similar will cluster together automatically
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
#scale will give you cluster 
```
