```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#make sure show on same website
```

```python
import numpy as np
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y,'red')
```

```python
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#00
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.5,0.15,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('small')
axes1.set_title('big')
```

```python
fig=plt.figure()#a blank canvas
#need fig.add_axes to add 
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
plt.tight_layout()#make it losely
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)
axes[1].set_title('second')
plt.tight_layout()#always do this to avoid overlap
#this is the second way of creating subplot
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('Xlabel')
```

```python
fig,axes=plt.subplots(nrows=2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#startmakeitcolorful
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,alpha=0.5,ls='steps',marker="o",
        markersize=10,markerfacecolor='yellow',markeredgewidth=3,
       markeredgecolor='green')#google to get 
#color:RGB HEX Code#linewith
#alpha:appha transparency
#lw:linewidth
#ls:linestyle,--,-:,:,steps
#marker shows ,+,o
```

```python
#two graphs at the same pics
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label="x")#add the inference
ax.plot(y,x,label="y")
ax.set_title('Title')
ax.set_xlabel('Xlabel')
ax.legend(loc=0)#loc=0means the best location
ax.legend(loc=(0.1,0.1))
```
